```python
import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt  
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import MinMaxScaler 
from sklearn.linear_model import LinearRegression  
from sklearn.metrics import r2_score 
from sklearn import svm
from sklearn.ensemble import RandomForestRegressor
```

```python
data = pd.read_csv('d:\jupyter\house price\Housing.csv')  
data
```

```python
data.isnull().sum()
```

```python
categorical_col =  ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

data[categorical_col]
```

```python
def binary_map(x):
    
    return x.map({'yes': 1, 'no': 0})
```

```python
data[categorical_col] = data[categorical_col].apply(binary_map)


data[categorical_col]
```

```python
dummy_col = pd.get_dummies(data['furnishingstatus'])


dummy_col.head()
```

```python
dummy_col = pd.get_dummies(data['furnishingstatus'], drop_first=True)

dummy_col.head()
```

```python
data = pd.concat([data, dummy_col], axis=1)


data.head()
```

```python
data.drop(['furnishingstatus'], axis=1, inplace=True)


data
```

```python
scaler = MinMaxScaler()

col_to_scale = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking', 'price']

data[col_to_scale] = scaler.fit_transform(data[col_to_scale])

data
```

```python
x = data.drop('price', axis=1)

y = data['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y , test_size=0.3, random_state=42)
```

```python
model_SVR = svm.SVR()

model_SVR.fit(x_train,y_train)

Y_pred = model_SVR.predict(x_test)


print("SVR Model R2:")
r2_score(y_test, Y_pred)
```

```python
model_RFR = RandomForestRegressor(n_estimators=10)

model_RFR.fit(x_train, y_train)

Y_pred = model_RFR.predict(x_test)

print("RF Model R2: ")
r2_score(y_test, Y_pred)
```

```python
model_LR = LinearRegression()

model_LR.fit(x_train,y_train)

Y_pred = model_LR.predict(x_test)


print("LR Model R2:")
r2_score(y_test, Y_pred)
```
